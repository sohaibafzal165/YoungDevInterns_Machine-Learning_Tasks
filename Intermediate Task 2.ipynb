## Removing Missing Values

```python
df['Distance from Home']=df['Distance from Home'].fillna(df['Distance from Home'].mean())
df['Company Tenure (In Months)']=df['Company Tenure (In Months)'].fillna(df['Company Tenure (In Months)'].mean())
```
