```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
```

```python
from sklearn.model_selection import cross_val_score, KFold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
from sklearn.metrics import make_scorer, accuracy_score, f1_score
```

```python
accuracy_scores = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')

f1_scores = cross_val_score(model, X, y, cv=kfold, scoring='f1_macro')

print("Accuracy scores:", accuracy_scores)
print("Mean Accuracy:", accuracy_scores.mean())
print("F1 scores:", f1_scores)
print("Mean F1 Score:", f1_scores.mean())
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, 6), accuracy_scores, marker='o', label='Accuracy')
plt.plot(range(1, 6), f1_scores, marker='s', label='F1 Score')
plt.xlabel('Fold')
plt.ylabel('Score')
plt.title('Cross-Validation Scores per Fold')
plt.legend()
plt.grid(True)
plt.show()
```
