```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
iris.target_names
```

```python
df.head()
```

```python
df['species'] = df['target'].map({i: name for i, name in enumerate(iris.target_names)})
```

```python
print(df['species'].value_counts())
```

```python
df.head()
```

```python
sns.scatterplot(
    data=df,
    x='petal length (cm)',
    y='petal width (cm)',
    hue='species',       # Color by species
    palette='Set1'
)

plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True)
plt.show()
```

```python
corr_matrix = df.iloc[:, :4].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Matrix (Iris Dataset)')
plt.show()
```

| Feature Pair              | Correlation  | Meaning                                       |
| ------------------------- | ------------ | --------------------------------------------- |
| **Petal length vs width** | \~ **0.96**  |  Strong positive correlation (very related) |
| **Sepal length vs width** | \~ **-0.12** |  Weak negative correlation (not related)    |
| **Sepal length vs petal** | \~ **0.87**  |  Good positive correlation                  |
